```python
import os
from ultralytics import YOLO

# Kaggle Dataset 自动挂载到 /kaggle/input/
DATA_YAML = "/kaggle/input/cghd-yolo/cghd_dataset.yaml"
MODEL_PT  = "/kaggle/input/cghd-v2-model/best.pt"

print(f"Dataset: {DATA_YAML} -> {os.path.exists(DATA_YAML)}")
print(f"Model:   {MODEL_PT} -> {os.path.exists(MODEL_PT)}")
```

```python
model = YOLO(MODEL_PT)
model.train(
    data=DATA_YAML,
    epochs=80,
    imgsz=640,
    batch=16,
    workers=2,
    resume=False,
    name='cghd_56cls_v3',
    project='/kaggle/working',
    lr0=0.0005, lrf=0.01,
    patience=15, warmup_epochs=3,
    cos_lr=True,
    hsv_h=0.015, hsv_s=0.7, hsv_v=0.4,
    degrees=10.0, translate=0.1, scale=0.5, shear=2.0,
    perspective=0.0005, flipud=0.5, fliplr=0.5,
    mosaic=1.0, mixup=0.1, erasing=0.4,
)
print('训练完成!')
```
